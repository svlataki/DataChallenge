We use a pretrained model which maps our documents to a 768 dimensional dense vector space. Sentence transformers are models based on transformer networks like BERT. The model we are using (https://huggingface.co/lighteternal/stsb-xlm-r-greek-transfer) is trained on greek sentences.

```python
import re
import zipfile
import pickle
from sentence_transformers import SentenceTransformer
```

```python

model = SentenceTransformer('lighteternal/stsb-xlm-r-greek-transfer')
```

```python
from google.colab import drive
drive.mount("/content/gdrive")
```

```python
text = dict()
with zipfile.ZipFile('/content/gdrive/My Drive/domains.zip', "r") as zfile:
    for filename in zfile.namelist():
        if re.search(r'\.zip$', filename) is not None:
            zfiledata = BytesIO(zfile.read(filename))
            with zipfile.ZipFile(zfiledata) as zfile2:
                text[filename[:-4]] = ''
                for name2 in zfile2.namelist():
                    file = zfile2.read(name2)
                    text[filename[:-4]] += file.decode('utf16') + ' '


data = [text[domain] for domain in text]
text=None
```

```python
embedding_features = []

for document in data:
    sentence_embeddings = model.encode(document)
    embedding_features.append(sentence_embeddings)
    
```

```python
with open('embedding_features.pkl', 'wb') as f:
    pickle.dump(embedding_features, f)
```
